```python
#el input de este programa es un directorio con archivos .csv y el output es el archivo file1csv.csv
import os
import csv
#se retiró β, se reemplazaron tʃʼ,tʃʰ,tʃ por "č","čʰ","č'" 
# se añadio "/" dado que ese caracter acompaña al formato IPA en los archivos de la carpeta csv
IPAquechua=["/","æ","ɑ","a","ɪ","i","e","ɛ","u","U","o","ʊ","ɔ","h","j","k","k̚","ɣ","x","kʰ","kx","k'","l","ʎ","m","n","ɲ","p","ɸ","pʰ","pɸ","p'","q","ɢ","X","qʰ","q'","ɾ","s","ʃ","t","tʰ","ts","θ","t'","č","čʰ","č'","W"]
NoAlFinalPalabra=["h","ñ","č'","čʰ","kʰ","k'","pʰ","p'","qʰ","q'","t","tʰ","t'"] 
alofonos=["ɛ","e","o","ɔ"]
uvular=["q","q'","qʰ"]
vowels=["æ","ɑ","a","ɪ","i","e","ɛ","u","U","o","ʊ","ɔ"]
consonants=["h","j","k","k̚","ɣ","x","kʰ","kx","k'","l","ʎ","m","n","ɲ","p","β","ɸ","pʰ","pɸ","p'","q","ɢ","X","qʰ","q'","ɾ","s","ʃ","t","tʰ","ts","θ","t'","č","čʰ","č'","W"]
directory='/media/alonso/3361-6630/csv/'
for filename in os.listdir(directory):
 with open(directory+filename, newline='') as csvfile:
  reader = csv.reader(csvfile)
  count=0
  for word,pron in reader:
    #verificar que la palabra examinada solamente contenga fonemas quechuas
    if all([char in IPAquechua for char in pron]): 
        #verificar que no haya consonantes prohibidas al final de palabra
        if pron[len(pron)-2] not in NoAlFinalPalabra:
          with open("file1csv.csv", "a") as f: 
            #verificar si la palabra tiene sonidos alofonos
            if all([char not in alofonos for char in pron]):
                for i in range(len(pron) - 1): # i is the current index
                    #verificar si la palabra tiene diptongos
                    if pron[i + 1] in vowels and pron[i] in vowels:
                        break
                    else:
                        if i == len(pron)-2:
                            for i in range(len(pron) - 1): # i is the current index
                                #verificar si la palabra tiene consonantes consecutivas
                                if pron[i + 1] in consonants and pron[i] in consonants:
                                    break
                                else:
                                    if i ==len(pron)-2:
                                        count+=1
                                        f.write(f"{count},{filename},{word},{pron}\n")
                                        #print(count,word,pron)
            else:
                #la palabra tiene alofonos, verificar si tiene uvulares 
                if any([char in uvular for char in pron]):
                    #la palabra tiene alofonos y uvulares, verificar si tiene diptongos
                    if pron[i + 1] in vowels and pron[i] in vowels:
                        break
                    else:
                        if i == len(pron)-2:
                            for i in range(len(pron) - 1): # i is the current index
                                if pron[i + 1] in consonants and pron[i] in consonants:
                                    break
                                else:  
                                    if i == len(pron)-2:
                                        count+=1
                                        f.write(f"{count},{filename},{word},{pron}\n")
                                        #print(count,word,pron)
```

```python
import os
import pandas as pd
DiccionarioFonemaGrafemaQuechua={"æ":"a","ɑ":"a","a":"a","ɪ":"i","i":"i","e":"i","ɛ":"i","u":"u","U":"u","o":"u","ʊ":"u","ɔ":"u","h":"h","j":"y","k":"k","k̚":"k","ɣ":"k","x":"k","kʰ":"kh","kx":"kh","k'":"k'","l":"l","ʎ":"ll","m":"m","n":"n","ɲ":"ñ","p":"p","ɸ":"p","pʰ":"ph","pɸ":"ph","p'":"p'","q":"q","ɢ":"q","X":"q","qʰ":"qh","q'":"q'","ɾ":"r","s":"s","ʃ":"sh","t":"t","tʰ":"th","ts":"th","θ":"th","t'":"t'","č":"ch","čʰ":"chh","č'":"ch'","W":"w"}
df=pd.read_csv("GlosaTSV.csv")
# la siguiente linea creó una columna, obviamente se ejecuta una sola vez
df.insert(loc=4,column="escrito en quechua",value="")
palabra_escrita_en_quechua=""
for index,word in enumerate(df["IPA"].values):
    for letter in word:
        if letter != " ":
        #buscamos en el diccionario el valor equivalente al caracter IPA
            palabra_escrita_en_quechua+=DiccionarioFonemaGrafemaQuechua[letter]
    #despues de construida la palabra quechua, la añadimos al dataframe, en la misma fila
    df.at[index,"escrito en quechua"]=palabra_escrita_en_quechua
    palabra_escrita_en_quechua=""
 #grabar en el archivo GlosaTSV
df.to_csv("GlosaTSV.csv")
```
